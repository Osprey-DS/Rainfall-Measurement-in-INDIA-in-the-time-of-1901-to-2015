## Importing Relevant Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.rcParams['figure.figsize'] = (12,8)
%matplotlib inline
```

```python
rain = pd.read_csv('district wise rainfall normal.csv', na_values=  0)
```

```python
rain.head(10)
```

```python
rain.shape
```

```python
state = rain['STATE_UT_NAME'].unique()

print(state.size)
state
```

```python
rain[rain['ANNUAL'] == rain['ANNUAL'].max()]
```

```python
rain[rain['ANNUAL'] == rain['ANNUAL'].min()]
```

```python
rain.groupby(by='STATE_UT_NAME').ANNUAL.mean()
```

```python
rain.groupby('STATE_UT_NAME').ANNUAL.agg(['count','mean' , 'min', 'max'])
```

```python

rain.groupby('STATE_UT_NAME').ANNUAL.agg(['mean' , 'min', 'max']).plot(kind='bar', figsize=(15,8))
plt.xlabel('STATE NAME')
plt.ylabel('COUNT')
plt.title('AGGREATE INFORMATION BY ANNUAL OF ALL STATE')
plt.tight_layout()
```
